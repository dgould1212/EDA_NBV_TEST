### 3. Check for potential relationships between continuous numerical features and price.

#### Create scatter plots of each continous numerical variable vs. price, and calculate the corresponding correlation coefficients and P-values.

```python
laptops.plot.scatter(x='Weight_kg', y='Price')
```

```python
cor_w, p_val_w = pearsonr(laptops['Weight_kg'], laptops['Price'])
cor_w, p_val_w
```

```python
laptops.plot.scatter(x='CPU_frequency', y='Price')
```

```python
cor_cpu, p_val_cpu = pearsonr(laptops['CPU_frequency'], laptops['Price'])
cor_cpu, p_val_cpu
```

```python
laptops.plot.scatter(x='Screen_Size_inch', y='Price')
```

```python
cor_s, p_val_s = pearsonr(laptops['Screen_Size_inch'], laptops['Price'])
cor_s, p_val_s
```
